```python

import pandas as pd
import json

df_fin = pd.DataFrame()

with open('/content/xaa') as f:
    df = pd.DataFrame(json.loads(line) for line in f)
with open('/content/xab') as f:
    df2 = pd.DataFrame(json.loads(line) for line in f)
with open('/content/xac') as f:
    df3 = pd.DataFrame(json.loads(line) for line in f) 
with open('/content/xad') as f:
    df4 = pd.DataFrame(json.loads(line) for line in f)
with open('/content/xae') as f:
    df5 = pd.DataFrame(json.loads(line) for line in f)
with open('/content/xaf') as f:
    df6 = pd.DataFrame(json.loads(line) for line in f)
with open('/content/xag') as f:
    df7 = pd.DataFrame(json.loads(line) for line in f)

list = [df, df2, df3, df4, df5, df6, df7]
for i in list:
  df_fin = df_fin.append(i[i.categories == "q-bio.NC"])
```
